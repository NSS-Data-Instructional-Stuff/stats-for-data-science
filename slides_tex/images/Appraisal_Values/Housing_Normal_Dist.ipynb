```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
houses = pd.read_csv('2017SINGLEFAMILYSF.csv')
```

```python
house_sample = houses.sample(n = 1000)
```

```python
house_sample = house_sample.rename(columns = {'2017 TOTAL APPR' : 'appraisal_value'})
```

```python
house_sample.appraisal_value.hist();
```

```python
house_sample.appraisal_value.apply(np.log).hist();
```

```python
import statsmodels.api as sm
import numpy as np
def make_qq_plot(data):
    mu = np.mean(data)
    sigma = np.std(data)
    
    sm.qqplot(data, line='45', loc = mu, scale = sigma);
```

```python
make_qq_plot(house_sample.appraisal_value);
```

```python
make_qq_plot(house_sample.appraisal_value.apply(np.log));
```

```python
mu = house_sample.appraisal_value.apply(np.log).mean()

sigma = house_sample.appraisal_value.apply(np.log).std()
```
