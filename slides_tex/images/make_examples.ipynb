```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from scipy.stats import norm
```

```python
def make_plot(area, direction = 'left', mu = None):
    if direction == 'left':
        x_min = -3
        x_max = norm.ppf(area)
    elif direction == 'right':
        x_min = norm.ppf(1-area)
        x_max = 3
    x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), 100)
    fig, ax = plt.subplots(1, 1, figsize = (8,4))
    ax.plot(x, norm.pdf(x),'r-', lw=3, alpha=1, label='norm pdf', color = 'black')
    if x_min < x_max:
        section = np.arange(x_min, x_max, 0.01)
        plt.fill_between(section,norm.pdf(section), color = 'slateblue')
    else:
        section_1 = np.arange(-3, x_max, 0.01)
        plt.fill_between(section_1, norm.pdf(section_1), color = 'slateblue')
        section_2 = np.arange(x_min, 3, 0.01)
        plt.fill_between(section_2, norm.pdf(section_2), color = 'slateblue')
    if x_min != -3:
        plt.vlines(x = x_min, ymin = 0, ymax = norm.pdf(x_min), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_min], labels = [f'z = {x_min}'], fontsize = 14)
        plt.annotate(s = f'?', xy = (x_min, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center', color = 'orangered')
        plt.annotate(s = f'Area = {round(1 - norm.cdf(x_min), 2)}', xy = ((3 + x_min)/2 - 0.5, 0.01), fontsize = 14,
                    va = 'bottom', ha = 'center', fontweight = 'bold')
    if x_max != 3:
        plt.vlines(x = x_max, ymin = 0, ymax = norm.pdf(x_max), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_max], labels = [f'z = {x_max}'], fontsize = 14)
        plt.annotate(s = f'?', xy = (x_max, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center', color = 'orangered')
        plt.annotate(s = f'Area = {round(norm.cdf(x_max), 2)}', xy = ((x_max-3)/2 + 0.5, 0.01), fontsize = 14,
                    va = 'bottom', ha = 'center', fontweight = 'bold')
    if mu:
        plt.xticks([0], labels = [mu], fontsize = 14, fontweight = 'bold')
    else:
        plt.xticks([])
    plt.hlines(y = 0, xmin = -3, xmax = 3)
    plt.yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.ylim(plt.ylim()[0] - 0.01, plt.ylim()[1]);
```

```python
make_plot(area=0.8, mu = 1000)
plt.savefig('ex_01.png')
```

```python
def make_plot_area(x_min = -3, x_max = 3, mu = 0, sd = 1):
    if x_min != -3:
        x_min_std = (x_min - mu) / sd
    else:
        x_min_std = x_min
    if x_max != 3:
        x_max_std = (x_max - mu) / sd
    else:
        x_max_std = x_max
    
    x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), 100)
    
    fig, ax = plt.subplots(1, 1, figsize = (8,4))
    ax.plot(x, norm.pdf(x),'r-', lw=3, alpha=1, label='norm pdf', color = 'black')
    if x_min_std < x_max_std:
        section = np.arange(x_min_std, x_max_std, 0.01)
        plt.fill_between(section,norm.pdf(section), color = 'slateblue')
    else:
        section_1 = np.arange(-3, x_max_std, 0.01)
        plt.fill_between(section_1, norm.pdf(section_1), color = 'slateblue')
        section_2 = np.arange(x_min_std, 3, 0.01)
        plt.fill_between(section_2, norm.pdf(section_2), color = 'slateblue')
    if x_min != -3:
        plt.vlines(x = x_min_std, ymin = 0, ymax = norm.pdf(x_min_std), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_min], labels = [f'z = {x_min}'], fontsize = 14)
        plt.annotate(s = f'{x_min}', xy = (x_min_std, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center')
    if x_max != 3:
        plt.vlines(x = x_max_std, ymin = 0, ymax = norm.pdf(x_max_std), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_max], labels = [f'z = {x_max}'], fontsize = 14)
        plt.annotate(s = f'{x_max}', xy = (x_max_std, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center')
    plt.hlines(y = 0, xmin = -3, xmax = 3)
    plt.yticks([])
    if mu != 0:
        plt.xticks([0], labels = [mu], fontsize = 14, color = 'red')
    else:
        plt.xticks([])    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.ylim(plt.ylim()[0] - 0.01, plt.ylim()[1]);
```

```python
make_plot_area(x_min = 1100, mu = 1000, sd = 200)
plt.savefig('ex_02.png')
```

```python
make_plot_area(x_min = 700, x_max = 900, mu = 1000, sd = 200)
plt.savefig('ex_03.png')
```

```python
make_plot_area(x_min = 1350, mu = 1000, sd = 200)
plt.savefig('ex_04.png')
```

```python
from ipywidgets import interact, FloatSlider
```

```python
style = {'description_width': 'initial'}
@interact(area = FloatSlider(value = 0.95, min = 0.6, max = .99, step = 0.01, description = "Confidence Level:", style = style),
    interval = FloatSlider(value = 0, min = -3, max = 3, step = 0.01, description = 'Sample Mean:', 
                           style = style, continuous_update = True))
def make_sym_plot(area = 0.95, annotations = False, interval = False):
    x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), 100)
    fig, ax = plt.subplots(1, 1, figsize = (14,4))
    ax.plot(x, norm.pdf(x),'r-', lw=3, alpha=1, label='norm pdf', color = 'black')
    x_min = norm.ppf((1-area) / 2)
    x_max = norm.ppf(1 - (1-area) / 2)
    if x_min < x_max:
        section = np.arange(x_min, x_max, 0.01)
        plt.fill_between(section,norm.pdf(section), color = 'slateblue')
    else:
        section_1 = np.arange(-3, x_max, 0.01)
        plt.fill_between(section_1, norm.pdf(section_1), color = 'slateblue')
        section_2 = np.arange(x_min, 3, 0.01)
        plt.fill_between(section_2, norm.pdf(section_2), color = 'slateblue')
    if x_min != -3:
        plt.vlines(x = x_min, ymin = 0, ymax = norm.pdf(x_min), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_min], labels = [f'z = {x_min}'], fontsize = 14)
        plt.annotate(s = f'{round(x_min,2)}', xy = (x_min, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center', color = 'orangered')
    if x_max != 3:
        plt.vlines(x = x_max, ymin = 0, ymax = norm.pdf(x_max), lw = 3, color = 'black')
        #plt.xticks(ticks = [x_max], labels = [f'z = {x_max}'], fontsize = 14)
        plt.annotate(s = f'{round(x_max,2)}', xy = (x_max, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center', color = 'orangered')
        
    plt.annotate(s = f'Area = {area}', xy = (0, 0.01), fontsize = 14,
                va = 'bottom', ha = 'center', fontweight = 'bold', color = 'yellow')
    
    if annotations:
        plt.annotate(s = f'Area = {round((1-area) / 2, 3)}', 
                     xy = (x_max + (3-x_max)/2, 0.05),
                     xytext = (max(3, x_max + 1), 0.15), 
                     arrowprops = {'color': 'black', 'width' : 3},
                     fontsize = 14, fontweight = 'bold', ha = 'right')
        plt.annotate(s = f'Area = {round((1-area) / 2, 3)}', 
                     xy = (x_min + (-3-x_min)/2, 0.05),
                     xytext = (min(-3, x_min - 1), 0.15), 
                     arrowprops = {'color': 'black', 'width' : 3},
                     fontsize = 14, fontweight = 'bold', ha = 'left')
        
    if True:
        plt.plot([interval + x_min, interval - 0.35], [-0.1, -0.1], lw = 3, color = 'red')
        plt.plot([interval + 0.35, interval + x_max], [-0.1, -0.1], lw = 3, color = 'red')
        plt.plot([interval + x_min, interval + x_min], [-0.09, -0.11], lw = 3, color = 'red')
        plt.plot([interval + x_max, interval + x_max], [-0.09, -0.11], lw = 3, color = 'red')
        plt.plot([interval, interval], [0, -0.08], lw = 3, linestyle = '--', color = 'black')
        plt.annotate(s = str(interval), xy = (interval, -0.1), 
                     va = 'center', ha = 'center',
                     fontsize = 14, fontweight = 'bold')
        plt.annotate(s = r'0', xy = (0, -0.01), fontsize = 14, fontweight = 'bold', 
                     va = 'top', ha = 'center', color = 'orangered')
        
    plt.plot([0, 0], [-0.15, 0.4], linestyle = '--', color = 'black')
        
    plt.hlines(y = 0, xmin = -3, xmax = 3)
    plt.yticks([])
    plt.xticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.ylim(plt.ylim()[0] - 0.01, plt.ylim()[1])
    plt.xlim(-5.5, 5.5);
```
